# Wikipedia Web Scraping and Data Cleaning
The following code are scrapes the wikipedia pages to get the movies that are in a film series together. It gathers the film series name, film movie name, year of release, and the wikipedia webpage link for the film.

```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import bs4
import requests
import re
import time
from sqlalchemy import create_engine
```

## Scrape List of Movie Sequels in Wikipedia

```python
wiki_list = ['two', 'three', 'four', 'five']
proxies = {'http' : 'http://206.189.157.23:80',
           'https' : 'https://206.189.157.23:80',}
for count in wiki_list:
    wiki_entry = requests.get('https://en.wikipedia.org/wiki/'
                    'List_of_feature_film_series_with_' + count + '_entries', 
                     proxies = proxies).text
    wiki_entry_soup = bs4.BeautifulSoup(wiki_entry)
    wiki_entry_result = []
    for series_name in wiki_entry_soup.select('.div-col > ul > li'):
        series = series_name.i.text
        for movie_name in series_name.select('ol > li '):
            try:
                year =  re.findall(r'\((\d{4})\)', movie_name.text)[0]              
            except:
                year = ""        
            try:
                html = ("https://en.wikipedia.org" 
                        + str(movie_name.i.a['href']))                   
            except:
                html = ""
            wiki_entry_result.append([series, count + " movies", 
                                      movie_name.i.text, year, html])
    df_wiki_entry = pd.DataFrame(wiki_entry_result, columns=['Movie Series', 
                                          'Movie Count', 'Movie Name', 
                                           'Year', 'Wikipedia Link'])
    df_wiki_entry.to_sql('Wikipedia_Movies', 
                         'sqlite:////mnt/processed/private/msds2021/'
                         'slt10c/Wikipedia_Movies.db', if_exists='append')
    time.sleep(np.random.random())
```

## Data Cleaning of Movie Sequels SQL Database

```python
def string_replace(string, pattern, replace):
    try:
        return re.sub(pattern, replace, string)
    except:
        return string
```

```python
conn = sqlite3.connect('/mnt/processed/private/msds2021/slt10c/'
                       'Wikipedia_Movies.db')
conn.execute('SELECT * FROM Wikipedia_Movies')
conn.create_function('REMOVE_STRING', 3, string_replace)
```
